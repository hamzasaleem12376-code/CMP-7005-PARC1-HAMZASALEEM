### **DISPLAYING THE MISSING VALUES**

```python
def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * mis_val / len(df)

    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False)

    return mis_val_table

missing_values = missing_values_table(df)
display(missing_values.style.background_gradient(cmap='coolwarm'))
```

```python
df = df.sort_values(['City', 'Date'])

numeric_cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'AQI']
df[numeric_cols] = df.groupby('City')[numeric_cols].ffill().bfill()

df['AQI_Bucket'] = df.groupby('City')['AQI_Bucket'].ffill().bfill()
```

I sorted the dataset by City and Date to ensure that missing values are filled in the correct chronological order within each city. Sorting the data first ensures that forward fill (ffill) and backward fill (bfill) work as intended.

Next, I applied both forward fill and backward fill to the numeric columns (such as PM2.5, PM10, NO, NO2, NOx, NH3, CO, SO2, O3, AQI) within each city. I chose to apply both of these techniques at the same time to make sure that the missing values are filled in a more robust and complete manner.

Why I Chose Both Forward Fill and Backward Fill:
Forward fill works by carrying the previous valid value forward. For example, if PM2.5 has a value of 30 for 2021-06-01 and is missing for 2021-06-02, forward fill will replace the missing value for 2021-06-02 with 30. This method works well when the data doesn't change drastically over short periods.
Backward fill, on the other hand, fills missing values using the next available value. For instance, if the PM2.5 value is missing on 2021-06-02 but the value for 2021-06-03 is 40, backward fill will replace the missing value for 2021-06-02 with 40. This ensures that even if ffill cannot fill the gap (like when there is no previous valid value), bfill can fill the missing data from subsequent values.

By using both techniques, I ensured that no missing value remains in the dataset, even if the gaps are at the start or end of the data.

For example:

If PM2.5 for 2021-06-02 is missing, but PM2.5 on 2021-06-01 is 30 (previous day) and 2021-06-03 is 40 (next day), the combination of both fills ensures that 2021-06-02 gets a sensible value.
Forward Fill will fill it with 30 (previous day value).
Backward Fill will fill it with 40 (next day value).

This combined approach allows me to keep the data complete and consistent, making it more reliable for analysis or modeling.

Finally, I applied the same process to the AQI_Bucket column, which categorizes the AQI into levels like "Satisfactory", "Moderate", etc. By filling missing values in the AQI_Bucket using both forward and backward techniques, I ensured that each city’s AQI categories are consistent with neighboring data.

# **Creating a Stacked Bar Chart to Compare Average Pollution Levels Across Cities**

```python

city_stats = df.groupby('City')[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'AQI']].mean().reset_index()

import plotly.express as px

# Melt the DataFrame to create a 'Pollutant' column for visualization
city_stats_melted = pd.melt(city_stats,
                             id_vars=['City'],  # Replace 'station' with 'City'
                             value_vars=['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'AQI'],  # Adjusted pollutant list
                             var_name='Pollutant',
                             value_name='Average Concentration')

# Create the stacked bar chart using Plotly
fig_city = px.bar(city_stats_melted,
                  x='City',  # Use 'City' as the x-axis
                  y='Average Concentration',
                  color='Pollutant',  # Color differentiation by pollutant
                  title='Average Pollution Levels by City (Stacked)',
                  labels={'City': 'Cities', 'Average Concentration': 'Average Concentration (µg/m³)'},
                  barmode='stack')

# Display the chart
fig_city.show()
```

```python
def max_polluted_cities(selected_pollutants):
    # Group the data by 'City' and calculate the average of the selected pollutant
    x1 = df[[selected_pollutants, 'City']].groupby(["City"]).mean().sort_values(by=selected_pollutants, ascending=False).reset_index()

    # Round the average pollutant concentration to two decimal places
    x1[selected_pollutants] = round(x1[selected_pollutants], 2)

    # Return the top 6 cities with the highest average concentration of the selected pollutant
    return x1[:6].style.background_gradient(cmap='OrRd')
```
